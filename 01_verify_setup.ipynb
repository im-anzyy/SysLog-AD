```python
import os

# Check if dataset files exist
raw_folder = "data/raw"
files = os.listdir(raw_folder)

print(" Files in data/raw:")
for f in files:
    size = os.path.getsize(os.path.join(raw_folder, f))
    print(f"   {f}  —  {size / (1024*1024):.2f} MB")
```
